```python
# this is a string so first i will have to convert it into list
import ast # due to this we can convert string into list
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):    # function defination , convert is a name of function
    L = []                 # this is a list in this we will strore the genres word
         
    for i in ast.literal_eval(obj):     # here with the help of ast.literal_eval( ) we can convert string into list
        L.append(i['name'])
    return L
       
     
     
    
```

```python
movies['genres'] = movies['genres'].apply(convert) # we call the convert function
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter = counter+1
        else:
            break
    return L  

 
      

            
            
                
```

```python
def fetch_movie_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L        
```

```python
movies['crew'] = movies['crew'].apply(fetch_movie_director)
```

```python
# now we want to merge name and surname in a single word basically we want to remove the space between name and surname so that our model does not get confused
 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x] )   
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x] )    
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x] )    
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x] )    

```

```python
# now we remove similar words(action, actions like) so what we are doing, we are deleting the similar word and containing only its root word like action

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```
